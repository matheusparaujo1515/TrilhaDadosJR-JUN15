| Variável                | Descrição

### Carregando os pacotes
É considerado uma boa prática carregar os pacotes que serão utilizados como a primeira coisa do programa.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Carregando os dados
O comando pd.read_csv é um comando da biblioteca pandas (pd.) e carrega os dados do arquivo csv indicado para um objeto dataframe do pandas

```python
# Carregando o dataframe
df = pd.read_csv('input/vendas_cursos_online.csv')
# Mostrar quantidade de linhas e colunas do dataframe
print(df.shape)
# Carregar as 5 primeiras linhas do dataframe
df
```

### Entendimento e transformação dos dados
Nesta etapa tipicamente avaliamos a distribuição de todas as variáveis.

```python
# Verificando os tipos de cada variável
df.info()
```

```python
# Transformando variável Data do tipo object para datetime
df['Data'] = pd.to_datetime(df['Data'])
df.info()
```

```python
# Verificando se há dados faltantes
df.isna().sum()
```

```python
# Verificando se há IDs de venda duplicados
df.duplicated('ID').sum()
```

```python
# Verificando quais os cursos mais venderam em diferentes datas
df['Nome do Curso'].value_counts()
```

### Plotagem de gráficos 

Nesta etapa, iremos plotar alguns gráficos das variáveis numéricas para avaliar a distribuição de todas elas. Vou criar um dataFrame separado, sem a coluna ``ID``, pois ela não carrega informações relevantes, já que é apenas um identificador.

```python
# Criando dataframe a parte e selecionando apenas as colunas númericas
df_sem_id = df.drop(columns = 'ID')
df_sem_id = df_sem_id.select_dtypes(include=['float64', 'int64'])
# Plotando pairplot com as variáveis númericas
sns.pairplot(df_sem_id)
plt.savefig('./output/pairplots.jpg')
```

```python
# Plotando gráfico de correlação
correlação = df_sem_id.corr()
sns.heatmap(correlação, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
plt.savefig('./output/Gráfico de correlação.jpg')
plt.show()
```

### Insight:
Após analisarmos os gráficos de dispersão, histogramas e a matriz de correlação, podemos notar que há uma correlação negativa forte (-0,86) entre ``Quantidade de Vendas`` e ``Preço Unitário``, o que sugere que, à medida que o preço unitário aumenta, a quantidade de vendas tende a diminuir e vice-versa.

### Resolução do desafio proposto

Nesta etapa, vamos abordar os seguintes desafios:

- Calcular a receita total gerada pela venda dos cursos.
- Identificar o curso com o maior número de vendas.
- Visualizar a distribuição das vendas ao longo do tempo através de gráficos

### Calcular a receita total gerada pela venda dos cursos.

Para calcularmos a receita, precisamos primeiro multiplicar a quantidade de vendas pelo preço unitário. Para isso, faremos uma cópia do dataFrame para então realizarmos essa operação.

```python
#Criando uma cópia
df1 = df.copy()
# Criando uma nova coluna 'Receita'
df1['Receita'] = df['Quantidade de Vendas'] * df['Preço Unitário']
df1.head()
```

```python
# Somando todas as receitas
receita_total = df1['Receita'].sum()
print(f'A receita total pela venda dos cursos foi de R${receita_total:.2f} reais.')
```

### Identificar o curso com o maior número de vendas.


```python
# Agrupando por Nome do curso e quantidade de vendas
df_group_curso_qtdvenda = df.groupby('Nome do Curso')['Quantidade de Vendas'].sum().reset_index()
# Colocando em ordem decrescente para visualizaros as maiores vendas
maiores_vendas = df_group_curso_qtdvenda.sort_values(by='Quantidade de Vendas', ascending=False)
maiores_vendas = maiores_vendas.reset_index(drop=True)
maiores_vendas.head()
```

```python
plt.figure(figsize=(10, 6))
maiores_vendas.plot(kind='bar', x='Nome do Curso', legend=False)
plt.xlabel('Cursos')
plt.ylabel('Quantidade de Vendas')
plt.title('Cursos com mais vendas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('./output/Cursos com mais vendas.jpg')
plt.show()
```

O curso com maior número de vendas foi "Introdução à Programação em Python" com 95 vendas.

### Visualizar a distribuição das vendas ao longo do tempo através de gráficos

```python
# Separando as variáveis Data e quantidade de vendas em grupos 
df_group_data_qtdvenda = df.groupby('Data')['Quantidade de Vendas'].sum().reset_index()
df_group_data_qtdvenda.head()
```

```python
# Plotando o gráfico
plt.figure(figsize=(15, 6))
plt.bar(df_group_data_qtdvenda['Data'], df_group_data_qtdvenda['Quantidade de Vendas'])

plt.xticks(rotation=90)
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.title('Quantidade de Vendas X Data')
plt.tight_layout()
plt.savefig('./output/Quantidade de Vendas X Data.jpg')
plt.show()
```

Acredito que também seja interessante acrescentar o gráfico de distribuição das receitas ao longo do tempo.

```python
# Separando as variáveis Data e quantidade de vendas em grupos 
df_group_data_receita = df1.groupby('Data')['Receita'].sum().reset_index()
df_group_data_receita.head()
# Plotando o gráfico
plt.figure(figsize=(15, 6))
plt.bar(df_group_data_receita['Data'], df_group_data_receita['Receita'])

plt.xticks(rotation=90)
plt.xlabel('Data')
plt.ylabel('Receita')
plt.title('Receita X Data')
plt.tight_layout()
plt.savefig('./output/Receita X Data.jpg')
plt.show()
```
